```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score
```

```python
data_test = pd.read_csv('titanic\\test.csv')
df = pd.read_csv ('titanic\\train.csv') #C = Cherbourg, Q = Queenstown, S = Southampton
```

```python
df['Age']=df['Age'].fillna(value=df['Age'].median())
```

```python
df['Sex_cat']=df['Sex'].factorize()[0] #konwertuje na macierz binarną i przypisuje do zmiennej kategorialnej
```

```python
#np.random.seed(9)
feats=['Pclass','Sex_cat','Fare','Age']
X = df[feats].values
y = df['Survived'].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
#model = DecisionTreeClassifier(max_depth=10)#moze być bez param max - puste
#model = RandomForestClassifier(max_depth=10,n_estimators=50 )
#model.fit(X_train, y_train)
#y_pred = model.predict(X_test)
#accuracy_score(y_test, y_pred)    #0.869 max
```

```python
kf = KFold(n_splits=10)
#kf.get_n_splits(X)
outcomes = []
fold = 0
for aaa in kf.split(X) :
    fold+=1
    model = RandomForestClassifier(max_depth=10,n_estimators=50 )
    model.fit(X_train,y_train)
    preds = model.predict(X_test)
    accur = accuracy_score(y_test,preds)
    outcomes.append(accur)
    print("fold {0} accuracy: {1}".format (fold,accur))
mean_outcome = np.mean(outcomes)
print ("mean accuracy:{0}".format(mean_outcome))
```
